```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (Top 10 countries by population in 2022)
data = {
    'Country': [
        'China', 'India', 'United States', 'Indonesia', 'Pakistan',
        'Nigeria', 'Brazil', 'Bangladesh', 'Russia', 'Mexico'
    ],
    'Population (2022)': [
        1411750000, 1407560000, 332000000, 275500000, 240500000,
        216700000, 214300000, 172900000, 144700000, 126000000
    ]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(12, 7))
bars = plt.barh(df['Country'], df['Population (2022)'], color='#1f77b4')
plt.xlabel('Population (in billions)', fontsize=12)
plt.title('Top 10 Most Populous Countries (2022)', fontsize=16, fontweight='bold')
plt.gca().invert_yaxis()

# Add labels
for bar in bars:
    plt.text(bar.get_width() + 10000000, bar.get_y() + bar.get_height()/2,
             f"{bar.get_width()/1e9:.2f}B", va='center', fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.style.use('seaborn-v0_8-darkgrid')
plt.savefig("top_10_population_2022.png", dpi=300)
plt.show()
```
